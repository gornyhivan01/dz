Этапы выполнения задания
Этап 1. Подготовка данных
Скачайте файл с набором данных событий информационной безопасности в формате JSON (events.json).
Важно: можно сгенерировать данные самостоятельно или использовать предоставленные.
 

Этап 2. Анализ данных
Загрузите данные из файла JSON в датафрейм с помощью Pandas и проанализируйте их. Вы должны изучить распределение событий информационной безопасности по их типам (по полю «signature»).
 

Этап 3. Визуализация данных
Постройте график распределения типов событий информационной безопасности. Используйте Seaborn или Matplotlib для создания визуализации.
 

```python
# Этап 1: Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import json

# Этап 2: Загрузка и подготовка данных
with open('events.json', 'r') as file:
    data = json.load(file)

df = pd.DataFrame(data['events'])

df.drop_duplicates(inplace=True)
df.replace({ "": pd.NA, "null": pd.NA, None: pd.NA}, inplace= True)
df.dropna(subset=["timestamp", "signature"], inplace=True)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors='coerce')
df["signature"] = df["signature"].str.strip()

# Анализ данных
df["hour"] = df["timestamp"].dt.hour
df["date"] = df["timestamp"].dt.date
df["day_of_week"] = df["timestamp"].dt.day_name()

# Визуализация данных
# ВИЗУАЛИЗАЦИЯ: Matplotlib (как в примере)
print("ВИЗУАЛИЗАЦИЯ: MATPLOTLIB")

# Используем Matplotlib для визуализации
plt.figure(figsize=(12, 8))
counts = df['signature'].value_counts()
counts.plot(kind='bar')
plt.title("Event Signatures Frequency")
plt.ylabel("Count")
plt.xlabel("Signature Name")
plt.xticks(rotation=45, ha="right")
plt.show()

# ВИЗУАЛИЗАЦИЯ: Seaborn countplot
print("\n" +"ВИЗУАЛИЗАЦИЯ: SEABORN COUNTPLOT")

import seaborn as sns
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='signature', order=df['signature'].value_counts().index)
plt.title("Event Signatures Frequency")
plt.ylabel("Signature Name")
plt.xlabel("Count")
plt.show()


# ВИЗУАЛИЗАЦИЯ: Распределение по типам событий
print("\n" +"ВИЗУАЛИЗАЦИЯ: РАСПРЕДЕЛЕНИЕ ТИПОВ СОБЫТИЙ")

# Создаем категории событий для лучшей визуализации
def get_event_type(signature):
    if 'MALWARE' in signature:
        return 'Malware'
    elif 'EXPLOIT' in signature:
        return 'Exploit'
    elif 'NETBIOS' in signature:
        return 'NetBIOS'
    elif 'INDICATOR' in signature:
        return 'Indicator'
    else:
        return 'Other'

df['event_type'] = df['signature'].apply(get_event_type)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="event_type") # Changed from "Тип события" to "event_type"
plt.title("Распределение типов событий безопасности")
plt.show()

# ВИЗУАЛИЗАЦИЯ: Корреляционная матрица (как в примере)
print("\n" + "ВИЗУАЛИЗАЦИЯ: КОРРЕЛЯЦИОННАЯ МАТРИЦА")

# Создаем числовые признаки для корреляционного анализа
df_analysis = df.copy()

# Кодируем категориальные переменные
df_analysis['signature_code'] = df_analysis['signature'].astype('category').cat.codes
df_analysis['event_type_code'] = df_analysis['event_type'].astype('category').cat.codes
df_analysis['day_code'] = df_analysis['day_of_week'].astype('category').cat.codes

# Создаем корреляционную матрицу
correlation_matrix = df_analysis[['hour', 'signature_code', 'event_type_code', 'day_code']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Корреляционная матрица")
plt.show()
```
