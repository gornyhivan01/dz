```python
# Импорт библиотек
import requests
import json

# Ключ API для VirusTotal
API_KEY = '2a1cf18c1e981b66431c43cb7ea8e9f520ad278fc1bb71557b6b99e7ea3e15ee'

# Базовый URL для VirusTotal API v3
BASE_URL = 'https://www.virustotal.com/api/v3'

# Заголовок для авторизации
headers = {
    'x-apikey': API_KEY
}

# Сканирование файла

file_path = 'temp.txt.'  # Файл должен находиться в той же директории

if not os.path.exists(file_path):
    print(f" Файл не найден: {file_path}")
else:
    print(f" Загрузка файла {file_path} для анализа...")

    # Запрос URL для загрузки
    upload_url = f"{BASE_URL}/files/upload_url"
    response = requests.get(upload_url, headers=headers)

    if response.status_code != 200:
        print(f"Ошибка получения URL для загрузки: {response.status_code} — {response.text}")
    else:
        upload_url = response.json()['data']

        # Загрузка файла
        with open(file_path, 'rb') as f:
            files = {'file': (os.path.basename(file_path), f)}
            response = requests.post(upload_url, headers=headers, files=files)

        if response.status_code == 200:
            result = response.json()
            file_id = result['data']['id']
            print(f"Файл успешно загружен. ID анализа: {file_id}")

            # Получение результатов анализа
            analysis_url = f"{BASE_URL}/analyses/{file_id}"
            print("Ожидание завершения анализа... (может занять время)")

            # # Повторяем запрос каждые 10 секунд до готовности
            import time
            while True:
                response = requests.get(analysis_url, headers=headers)
                if response.status_code == 200:
                    report = response.json()
                    status = report['data']['attributes']['status']
                    if status == 'completed':
                        stats = report['data']['attributes']['stats']
                        print("Результаты анализа файла:")
                        print(json.dumps(stats, indent=2))
                        break
                    elif status == 'queued' or status == 'in-progress':
                        print("Анализ ещё в очереди или выполняется... Ждём 10 секунд.")
                        time.sleep(10)
                    else:
                        print(f"Неизвестный статус: {status}")
                        break
                else:
                    print(f"Ошибка при получении отчёта: {response.status_code}")
                    break
        else:
            print(f"Ошибка загрузки файла: {response.status_code} — {response.text}")


# Запуск:
# Убедитесь, что установлены зависимости: pip install requests json
# Поместите файл 'temp.txt' в ту же папку, где находится скрипт.
# Запустите скрипт.
# API-ключ задать напрямую.
# Вывод по файлу в консоль
```
