Задание 1
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
The Moscow Times - Wednesday, October 2, 2002
The Guardian - Friday, 11.10.13
Daily News - Thursday, 18 August 1977

Пример работы программы
Программа должна выводить на экран объекты типа datetime, соответствующие датам в условии задачи
 
 

```python
from datetime import datetime

date_moscow_times_str = "Wednesday, October 2, 2002"
date_guardian_str = "Friday, 11.10.13"
date_daily_news_str = "Thursday, 18 August 1977"

dt_moscow_times = datetime.strptime(date_moscow_times_str, "%A, %B %d, %Y")

dt_guardian = datetime.strptime(date_guardian_str, "%A, %d.%m.%y")

dt_daily_news = datetime.strptime(date_daily_news_str, "%A, %d %B %Y")

print(f"The Moscow Times: {dt_moscow_times}")
print(f"The Guardian: {dt_guardian}")
print(f"Daily News: {dt_daily_news}")
```
