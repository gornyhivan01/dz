Используйте API OpenWeather для получения данных о погоде.
Напишите программу, которая:
· принимает название города от пользователя,
· отправляет GET-запрос к API и выводит текущую температуру и описание погоды.

```python
import requests

response = requests.get("https://jsonplaceholder.typicode.com/posts") #запрос к публичному API

if response.status_code == 200:
    posts = response.json()

 #   print (f" {posts [0:1]} \n")
    

    for post in posts[:5]:
        print(f"Номер: {post['id']}")
        print(f"Заголовок: {post['title']}")
        print(f"Что внутри: {post['body']}")
        print("--------------------")
else:
    print(f"Ошибка при выполнении запроса: {response.status_code}")
```

```python
import requests

def get_weather(city_name, api_key):
    
    url = "https://api.openweathermap.org/data/2.5/weather"
    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric',  #Температура в градусах Цельсия
        'lang': 'ru'        #На русском языке
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        
        #print (data)
        
        country = data['sys']['country'] #Страна
        city = data['name'] #Город
        temp = data['main']['temp'] #Температура
        vl = data['main']['humidity'] #Влажность
        description = data['weather'][0]['description'] #Погода
        
        print(f"Город: {city}, {country}")
        print(f"Текущая температура: {temp}°C")
        print(f"Влажность: {vl}%")
        print(f"Погода: {description}")
    else:
        try:
            error_message = response.json().get('message', 'Неизвестная ошибка')
        except:
            error_message = 'Ошибка подключения или неверный ответ от сервера'
        print(f" Не удалось получить данные: {error_message}")

#Запрос города
city = input("Введите название города: ")
get_weather(city, API_KEY)
```
