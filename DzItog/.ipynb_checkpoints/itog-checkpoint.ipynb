```python
# Установка необходимых библиотек
import subprocess
import re
import matplotlib.pyplot as plt
import json
import requests
from datetime import datetime
import os

# Парсинг лог-файла export.txt лежит в этой же директории (формат Suricata)
log_file = 'export.txt'

if not os.path.exists(log_file):
    raise FileNotFoundError(f" Файл {log_file} не найден.")

print(f" Загружаем и парсим лог: {log_file}")

# Регулярное выражение для строк лога
pattern = re.compile(
    r"(\d{2}\.\d{2}\.\d{4} \d{2}:\d{2}:\d{2})\s*\|\s*\[(\w+)\]\s*\[(\d+:\d+:\d+)\]\s*([^\[]+?)\s*"
    r"\[Classification:\s*([^\]]+)\]\s*\[Priority:\s*(\d)\]\s*\{(\w+)\}\s*"
    r"(\d+\.\d+\.\d+\.\d+):(\d+)\s*->\s*(\d+\.\d+\.\d+\.\d+):(\d+)"
)

# Счётчики приоритетов
priority_stats = {
    1: {'count': 0, 'ips': []},
    2: {'count': 0, 'ips': []},
    3: {'count': 0, 'ips': []}
}

# Собираем IP с Priority 1
high_priority_ips = []

with open(log_file, 'r', encoding='utf-8') as f:
    for line_num, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue
        match = pattern.match(line)
        if match:    
            # Извлекаем данные
            _, _, _, _, _, priority_str, _, src_ip, _, _, _ = match.groups()
            priority = int(priority_str)

            if priority in priority_stats:
                priority_stats[priority]['count'] += 1
                priority_stats[priority]['ips'].append(src_ip)

                # Сохраняем IP с Priority 1
                if priority == 1:
                    high_priority_ips.append(src_ip)
        else:
            print(f" Не удалось распарсить строку {line_num}: {line[:100]}...")

print(f" Парсинг завершён.")

# Подсчёт IP с высоким приоритетом
if high_priority_ips:
    ip_freq = {}
    for ip in high_priority_ips:
        ip_freq[ip] = ip_freq.get(ip, 0) + 1
    top_attackers = sorted(ip_freq.items(), key=lambda x: x[1], reverse=True)[:5]
else:
    top_attackers = []

# Подсчет угроз по приоритетам
high_count = priority_stats[1]['count']
medium_count = priority_stats[2]['count']
low_count = priority_stats[3]['count']

print("\n Статистика по приоритетам:")
print(f"Высокий   (1): {high_count}")
print(f"Средний   (2): {medium_count}")
print(f"Низкий    (3): {low_count}")

if high_count > 0:
    print(f"\n Найдено {high_count} событий с ВЫСОКИМ приоритетом! Требуется немедленное внимание!")
elif medium_count > 0:
    print(f"\n {medium_count} событий со СРЕДНИМ приоритетом. Мониторим дальше.")
else:
    print("\n Все события — низкого приоритета. Система в норме.")


# 5 IP  с высоким приоритетом
if top_attackers:
    print("\n 5 IP с высоким приоритетом:")
    for ip, count in top_attackers:
        print(f"   {ip}: {count} раз")
else:
    print("\n Нет событий с высоким приоритетом для анализа IP.")


# Проверка IP через VirusTotal
VT_API_KEY = '2a1cf18c1e981b66431c43cb7ea8e9f520ad278fc1bb71557b6b99e7ea3e15ee'
VT_BASE_URL = 'https://www.virustotal.com/api/v3/ip_addresses'

def check_ip_virustotal(ip):
    url = f"{VT_BASE_URL}/{ip}"
    headers = {'x-apikey': VT_API_KEY}
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            data = response.json()
            stats = data['data']['attributes']['last_analysis_stats']
            return {
                'ip': ip,
                'malicious': stats.get('malicious', 0),
                'suspicious': stats.get('suspicious', 0),
                'undetected': stats.get('undetected', 0),
                'total': sum(stats.values()),
                'error': None
            }
        else:
            return {'ip': ip, 'error': f"HTTP {response.status_code}", 'malicious': 0, 'suspicious': 0, 'undetected': 0, 'total': 0}
    except requests.RequestException as e:
        return {'ip': ip, 'error': str(e), 'malicious': 0, 'suspicious': 0, 'undetected': 0, 'total': 0}

# Проверяем IP с высоким приоритетом
if top_attackers:
    suspicious_ips = [ip for ip, _ in top_attackers]
    print(f"\n Проверка подозрительных IP через VirusTotal")

    vt_results = []
    for ip in suspicious_ips:
        print(f"  Проверка: {ip}")
        result = check_ip_virustotal(ip)
        vt_results.append(result)
else:
    vt_results = []
    print(f"\n Нет подозрительных IP для проверки через VirusTotal.")

# Формируем данные для таблицы и рекомендации

table_data = []

for result in vt_results:
    ip = result['ip']
    malicious = result['malicious']
    suspicious = result['suspicious']
    undetected = result['undetected']
    total = result['total']
    error = result['error']

    if error:
        status = " Ошибка"
        rec = error
    elif malicious >= 10:
        status = " Высокий риск"
        rec = "Немедленно заблокировать."
    elif malicious >= 5:
        status = " Средний риск"
        rec = "Временная блокировка."
    elif malicious > 0:
        status = " Низкий риск"
        rec = "Проверить активность."
    else:
        status = " Без угроз"
        rec = "Рекомендации не требуются."

    table_data.append({
        'IP': ip,
        'Mal': malicious,
        'Sus': suspicious,
        'Und': undetected,
        'Статус': status,
        'Рекомендации': rec
    })

# Вывод таблицы
if table_data:
    
    print("\n РЕЗУЛЬТАТЫ ПРОВЕРКИ IP ЧЕРЕЗ VIRUSTOTAL И РЕКОМЕНДАЦИИ\n")

    # Заголовки
    print(f"{'IP':<15} {'Malicious':<14} {'Suspicious':<14} {'Undetected' :<14} {'Статус':<15} {'Рекомендации'}")
    print("-" * 105)

    # Данные
    for row in table_data:
        print(f"{row['IP']:<15} {row['Mal']:<14} {row['Sus']:<14} {row['Und']:<14} {row['Статус']:<15} {row['Рекомендации']}")
else:
    print(" Нет данных для отображения таблицы.")


# Круговая диаграмма по количеству приоритетов
labels = ['Высокий (1)', 'Средний (2)', 'Низкий (3)']
sizes = [high_count, medium_count, low_count]
colors = ['#d62728', '#ff7f0e', '#2ca02c']  # Красный, оранжевый, зелёный
explode = (0.1, 0, 0)  # Выделяем сектор "Высокий"

if sum(sizes) > 0:
    plt.figure(figsize=(9, 9))
    wedges, texts, autotexts = plt.pie(
        sizes,
        labels=labels,
        autopct=lambda pct: f'{pct:.1f}%\n({int(pct / 100 * sum(sizes))})',
        startangle=90,
        colors=colors,
        explode=explode,
        textprops={'fontsize': 12}
    )
    plt.title('\n Распределение событий по уровням приоритета', fontsize=16, pad=20)
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig('priority_chart.png', dpi=100)
    plt.show()
    print(" Круговая диаграмма 'priority_chart.png' сохранена.")
else:
    print(" Нет данных для построения круговой диаграммы.")


# Формирование отчёта
report = {
    "summary": {
        "total_high_priority": high_count,
        "total_medium_priority": medium_count,
        "total_low_priority": low_count,
    },
    "top_attackers_with_priority_1": [{"ip": ip, "count": count} for ip, count in top_attackers],
    "virustotal_check": [
        {k: v for k, v in r.items() if k != 'error'}
        for r in vt_results if not r.get('error')
    ],
    "recommendations": [
        {row['IP']: row['Рекомендации']} for row in table_data
    ]
}

# Сохранение в JSON
output_json = 'analysis_report.json'
with open(output_json, 'w', encoding='utf-8') as f:
    json.dump(report, f, indent=2, ensure_ascii=False)
print(f" Отчёт сохранён в: {output_json}")

print("\n Анализ завершён. Все результаты готовы.")
```
